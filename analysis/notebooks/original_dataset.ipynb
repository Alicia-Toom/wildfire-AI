```python
import os

import matplotlib.pyplot as plt
import numpy as np
import random as r

from PIL import Image
```

```python
def plot_samples(images: list, title: str):
    fig, axes = plt.subplots(
            nrows=3,
            ncols=3,
            figsize=(8, 8),
            subplot_kw={'xticks': [], 'yticks': []}
        )

    for i, ax in enumerate(axes.flat):
        image = images[i]
        ax.imshow(image)

    fig.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()
```

## DATASET

We took the dataset and sorted the images into __fire__, __non-fire__ and __satellite__ images.<br>
The CNN-model for the *Image Classification* needs the images to be in the same input-size, so we  resized the images to 250x250 pixels.

```python
ROOT = 'c:/code/projects/python/applied-ai'  # Changes this locally
DATA_PATH = str(ROOT + '/data/new-data/datasets/full/')
```

#### Number of Images

```python
fire_images = os.listdir(DATA_PATH + 'fire')
nonfire_images = os.listdir(DATA_PATH + 'no-fire')
satellite_images = os.listdir(DATA_PATH + 'satellite')

len(fire_images)
```

```python
X = ['FIRE', 'NON FIRE', 'SATELLITE']
y = [len(fire_images), len(nonfire_images), len(satellite_images)]
cmap = plt.get_cmap('viridis')
rescale = lambda y: (y - np.min(y)) / (np.max(y) - np.min(y))

fig = plt.figure(figsize=(5,5))

plt.bar(X, y, color=cmap.colors, width=0.4)
plt.xlabel('Image Class')
plt.ylabel('No. of Images')

fig.suptitle('Number Of Images Per Class')
plt.show()
```
